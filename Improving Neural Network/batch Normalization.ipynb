```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('concertriccir2.csv')
```

```python
df.columns = ['x', 'y', 'class']
df.head()
```

```python
plt.figure()
plt.scatter(df['x'], df['y'], c=df['class'])
plt.show()
```

```python
x = df[['x', 'y']].values
y = df['class'].values
```

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import BatchNormalization
```

```python
model = Sequential()

model.add(Dense(3, input_dim=2, activation='relu'))
model.add(Dense(2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=200, validation_split=0.2)
```

```python
model1 = Sequential()

model1.add(Dense(3, input_dim=2, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(2, activation='relu'))
model1.add(BatchNormalization())
model1.add(Dense(1, activation='sigmoid'))

model1.summary()
```

```python
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history1 = model1.fit(x, y, epochs=200, validation_split=0.2)
```

```python
plt.figure()
plt.plot(history.history['val_accuracy'])
plt.plot(history1.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['no batch normalization', 'batch normalization'], loc='upper left')
plt.show()
```
