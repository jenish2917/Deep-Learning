```python
import pandas as ps
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = ps.read_csv('DailyDelhiClimate.csv')
```

```python
df = df[['date', 'meantemp']]
df
```

```python
plt.scatter(df['date'], df['meantemp'], c ='r')
```

```python
x1 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
x1
```

```python
df['ewm'] = x1
```

```python
df
```

```python
plt.scatter(df['date'], df['meantemp'], c ='r')
plt.plot(df['date'], x1, c ='b', lw = 1, label = 'alpha = 0.9')
plt.legend()
plt.show()
```
